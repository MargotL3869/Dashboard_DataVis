```python
import cdsapi

MON_URL = "https://cds.climate.copernicus.eu/api"
MA_CLE = "e6879404-6927-41a8-8854-16f764f95c0d"

c = cdsapi.Client(url=MON_URL, key=MA_CLE)

print("D√©coupage en petites tranches de 5 ans pour passer sous le radar...")

# On avance de 5 en 5 (1950, 1955, 1960...)
for annee_debut in range(2025, 2026, 1):
    # L'ann√©e de fin est 4 ans plus tard (ex: 1950 + 4 = 1954)
    annee_fin = min(annee_debut + 4, 2025)

    # On cr√©e la liste des ann√©es pour cette tranche pr√©cise
    annees_tranche = [str(y) for y in range(annee_debut, annee_fin + 1)]
    nom_fichier = f'meteo_france_{annee_debut}_{annee_fin}.nc'

    print(f"‚è≥ T√©l√©chargement de la p√©riode {annee_debut} - {annee_fin}...")

    c.retrieve(
        'reanalysis-era5-single-levels',
        {
            'product_type': 'reanalysis',
            'format': 'netcdf',
            'variable': ['2m_temperature', 'total_precipitation'],
            'year': annees_tranche,
            'month': [str(m).zfill(2) for m in range(1, 13)],
            'day': [str(d).zfill(2) for d in range(1, 32)],
            'time': ['00:00', '06:00', '12:00', '18:00'],
            'area': [51.5, -5.5, 41.0, 10.0],
        },
        nom_fichier
    )
    print(f"‚úÖ {nom_fichier} t√©l√©charg√© avec succ√®s !")

print("üéâ FINI ! Vous avez triomph√© des serveurs de Copernicus !")
```
